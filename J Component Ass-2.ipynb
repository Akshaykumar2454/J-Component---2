```python
import pandas as pd
b=pd.read_csv('/content/Salary_Data-1 (1).csv')
print(b)
```

```python
x1=b.iloc[:,0:1]
y1=b.iloc[:,1:2]
print(x1,y1)
```

```python
from sklearn.model_selection import train_test_split
X_train,x_test,y_train,y_test=train_test_split(x1,y1,test_size=0.10,random_state=5)
print(X_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
import pandas as pd
import numpy as np
b=pd.read_csv('/content/Salary_Data-1 (1).csv')
from sklearn.linear_model import Ridge 
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
rr= Ridge(alpha=0.01)
rr.fit(X_train, y_train) 
pred_train_rr= rr.predict(X_train)
print(np.sqrt(mean_squared_error(y_train,pred_train_rr)))
print(r2_score(y_train, pred_train_rr))
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error,r2_score
from sklearn.linear_model import Lasso 
model_lasso= Lasso(alpha=0.01)
model_lasso.fit(x1,y1) 
pred_train_lasso= model_lasso.predict(x_test)
print(mean_squared_error(y_test,pred_train_lasso))
```
